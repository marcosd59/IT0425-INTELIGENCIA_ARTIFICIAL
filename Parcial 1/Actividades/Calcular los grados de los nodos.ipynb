```python
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
def create_graph(num_nodes):
	# Crear un grafo vacío
	G = nx.Graph()

	# Añadir nodos al grafo
	for i in range(1, num_nodes + 1):
		G.add_node(i)

	# Conectar cada nodo con otros nodos seleccionados al azar
	for node in G.nodes():
		# Número de conexiones aleatorias para cada nodo
		num_connections = random.randint(1, num_nodes - 1)

		# Elegir nodos aleatorios para conectar
		connections = random.sample(list(G.nodes()), num_connections)

		# Añadir aristas al grafo
		for connection in connections:
			if node != connection and not G.has_edge(node, connection):
				G.add_edge(node, connection)

	# Dibujar el grafo
	# Usa un layout diferente para una mejor visualización
	pos = nx.spring_layout(G)
	nx.draw(G, pos, with_labels=True, node_color="skyblue", edge_color="gray")

	# Mostrar el gráfico
	plt.show()

	# Mostrar el grado de cada nodo
	for node in G.nodes:
		print(f"Nodo {node}: Grado {G.degree[node]}")
```

```python
# Número de nodos para el grafo
num_nodes = 8

# Llamar a la función con el número de nodos ingresado
create_graph(num_nodes)
```
